# Importation des bibliotheques necessaires

```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
import os
from google.colab import files
```

# Traitement d'augmentation de luminosité d'image

```python
def increase_brightness(image, brightness):
  gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
  bright_image = gray_image + brightness
  bright_image = np.clip(bright_image, 0, 255)
  bright_image = cv2.cvtColor(bright_image, cv2.COLOR_GRAY2BGR)
  return bright_image
```

# Détection des contours avec l'opérateur de Canny

```python
def detect_edges_canny(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    edges = cv2.Canny(gray, 50, 150)  # Les valeurs 50 et 150 sont des seuils min et max

    return edges
```

# Chargement, traitement et téléchargement d'images

```python
image = cv2.imread("/content/part0.png")
cv2_imshow(image)

# Appliquer vos fonctions sur l'image
image1 = increase_brightness(image, 100)
image2 = detect_edges_canny(image1)
cv2_imshow(image2)

# Enregistrer l'image dans le répertoire actuel
image_path = "part1.png"
cv2.imwrite(image_path, image2)
print(f"L'image a été enregistrée : {image_path}")

# Télécharger l'image sur votre bureau à partir de Colab
files.download(image_path)
```
